#2 Explain the difference between mutable and immutable data types with examples?
   - Ans2:Mutable data types (e.g., lists) can change after creation, while immutable types (e.g., tuples) cannot be modified once created.
#3 What are the main differences between lists and tuples in Python?
  -Ans3:Lists are mutable, slower, and use more memory, while tuples are immutable, faster, and consume less memory in Python.
#4 Describe how dictionaries store data.
   -Ans4:Dictionaries store data as key-value pairs using a hash table, enabling efficient data retrieval based on unique, immutable keys.
#5 Why might you use a set instead of a list in Python?
   -And 5:Use a set over a list for unique elements, faster membership checks, and operations like unions, intersections, and differences in Python.

#6 What is a string in Python, and how is it different from a list?
    -Ans6:A string is an immutable sequence of characters, while a list is mutable and can store multiple data types in Python.
#7 How do tuples ensure data integrity in Python?
   -And 7:Tuples ensure data integrity by being immutable, meaning their contents cannot be altered after creation, protecting data from accidental modifications.
#8 What is a hash table, and how does it relate to dictionaries in Python?
   -Ans 8:A hash table stores data using hash functions for fast lookups. Python dictionaries use hash tables to manage key-value pairs efficiently.

#10 Explain why strings are immutable in Python?
   -Ans 10:Strings are immutable in Python to optimize memory usage and performance, ensure thread safety, and allow strings to be used as keys in hash-based structures like dictionaries.
#11 What advantages do dictionaries offer over lists for certain tasks?
   -Ans11:Dictionaries offer faster lookups, insertions, and deletions due to their hash table implementation, making them more efficient than lists for tasks involving key-based data retrieval.
#12 Describe a scenario where using a tuple would be preferable over a list?
    -Ans12:Using a tuple is preferable when storing fixed, unchangeable data, such as coordinates (x, y, z), where data integrity and immutability are important.
#13 How do sets handle duplicate values in Python?
    -Ans13:Sets automatically remove duplicate values, ensuring that only unique elements are stored. Any repeated elements are discarded upon insertion.
#14 How does the “in” keyword work differently for lists and dictionaries?
   -Ans 14:For lists, "in" checks for element presence; for dictionaries, it checks for key existence, not values or items.

   -Ans15:No, tuples are immutable, meaning their elements cannot be modified after creation, ensuring data integrity and consistency.

#17 Describe the time complexity of accessing elements in a dictionary?
   -Ans17:Accessing elements in a dictionary has an average time complexity of O(1), due to hash-based key lookups.
#18 In what situations are lists preferred over dictionaries?
   -Ans18:Lists are preferred when order matters, or when handling a collection of items without needing fast key-based lookups or unique elements.
#19 Why are dictionaries considered unordered, and how does that affect data retrieval?
   -Ans 19:Dictionaries are unordered because keys are stored in a hash table, affecting retrieval by key but not by insertion order.
#20 Explain the difference between a list and a dictionary in terms of data retrieval.
